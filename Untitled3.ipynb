```python
# Load your dataset
data = pd.read_csv("/content/drive/MyDrive/Salaries.csv")
```

```python
# check for missing values in each column
print(data.isnull().sum())
data.isnull()
```

```python
# Calculate basic statistics mean, median, minimum, and maximum salary, and find the standard deviation
data.describe()
```

```python
#  determine the range of salaries
print(data.max(numeric_only=True) - data.min(numeric_only=True))
```

```python
# Calculate the mode
data.mode(numeric_only=True)
```

```python
# Handle missing data
data.drop(["Status", "Notes"], inplace = True, axis = 1)  # drop those columns because all thir data are null
data.dropna(axis=0, how='all')
```

```python
for x in data.index:
  for y in data.columns:
    if isinstance(data.loc[x, y], (int, float)) and data.loc[x, y] < 0:
      data.loc[x, y] = 0;
```

```python
data.fillna(data.median(), inplace=True)
```

```python
print(data.isnull().sum())
```

```python
sns.histplot(data=data, x='BasePay');
```
